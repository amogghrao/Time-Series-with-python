## Histogram
[Histogram](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.hist.html#pandas.DataFrame.hist) groups numbers into ranges/bins. The height of each bar shows how many fall into each range and width is defined by size of ranges/bins. 

```python
df1['A'].plot.hist()
```

We need to utilize the parameter `edgecolor` to create distinct bars,basically adding edges to the vertical bins.The possible values for edge colors can be found in one of `'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'`

```python
df1['A'].plot.hist(edgecolor ='k')
```

Simple axis view autoscaling. It turns autoscaling on or off, and then, if autoscaling for either axis is on, it performs the autoscaling on the specified axis or axes.[autoscale](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.autoscale.html)

```python
df1['A'].plot.hist(edgecolor ='k').autoscale(enable = True,axis = 'both',tight = True)
```

We can also change the number of bins (the range of values over which frequencies are calculated) from the default value of 10:

```python
df1['A'].plot.hist(edgecolor ='k',bins = 20).autoscale(enable = True,axis = 'both',tight = True)
```

To add grid to the graph. Use the parameter `grid`

```python
df1['A'].plot.hist(edgecolor ='k',grid = True,bins = 20).autoscale(enable = True,axis = 'both',tight = True)
```

Read through the parameter list of `df.plot.hist` to explore other customization offered.

## Barplots

A [barplot](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.bar.html) (or barchart) is one of the most common type of plot. It shows the relationship between a numerical variable and a categorical variable.For example, you can display the height of several individuals using bar chart.Barcharts are often confused with histograms, which is highly different. (It has only a numerical variable as input and shows its distribution).

```python
df2
```

```python
df2.plot.bar() #plots all index column as categorical values and rest of them as numercal values 
```

```python
df2.plot.bar(stacked = True)
```

```python
df2.plot.barh(stacked = True)
```

## Line Plots
Line plots are used to compare two or more variables. By default, the x-axis is the index of the dataframe.

Line plots is the default pandas plot. They are accessible through `df.plot()` as well as `df.plot.line()`

```python
df2
```

```python
df2.plot.line(y = 'a') #plotting 'a' column on the y axis with index as x-axis
```

If we want to change the figure size,make it bigger or smaller. we use `figsize` parameter of `line` function. It is passed as tuple and it is size of x and y axis in inches.

```python
df2.plot.line(y = 'a',figsize = (14,4)) #14" by 4"
```

If we want to change the line width,make it bigger or smaller. we use `lw` parameter of `line` function.

```python
df2.plot.line(y = 'a',figsize = (14,4),lw = 4) 
```

```python
#Multiple series
df2.plot.line(y=['a','b','c'],figsize = (18,4),lw = 4)
```

## Area Plots
A stacked area chart is the extension of a basic area chart to display the evolution of the value of several groups on the same graphic. The values of each group are displayed on top of each other. It allows to check on the same figure the evolution of both the total of a numeric variable, and the importance of each group.[reference](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html#pandas.DataFrame.plot)



```python
df2
```

```python
df2.plot.area(figsize = (18,5))
```

 too bright? increase transparency by passing `alpha` parameter which ranges between 0 to 1. 0 being completly transparent, 1 being opaque

```python
df2.plot.area(figsize = (18,5),alpha = 0.5)
```

```python
#don't want it stacked?
df2.plot.area(figsize = (18,5),alpha = 0.5,stacked = False)
```

## Scatter Plots
A [Scatterplot](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.scatter.html) displays the value of 2 sets of data on 2 dimensions. Each dot represents an observation. The position on the X (horizontal) and Y (vertical) axis represents the values of the 2 variables

```python
df1.head()
```

```python
df1.plot.scatter(x = 'A',y = 'B')
```

You can use `c` parameter to color each marker based off another column value. Use `cmap` to indicate which colormap to use.<br>
For all the available colormaps, check out: http://matplotlib.org/users/colormaps.html

```python
df1.plot.scatter(x = 'A',y = 'B',c = 'C',cmap = 'coolwarm')
```

Alternatively you can use `s` parameter to indicate marker size based off another column. The `s` parameter needs to be an array, not just the name of a column:

```python
df1.plot.scatter(x = 'A',y = 'B',s = df1['C']*100,alpha = 0.5,figsize = (18,12))
```

## KDE plots
[KDE Plot](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.kde.html) described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable.

## Hexagonal Bin Plot
[Hexagonal Binning](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.hexbin.html) is another way to manage the problem of having to many points that start to overlap. Hexagonal binning plots density, rather than points. Points are binned into gridded hexagons and distribution (the number of points per hexagon) is displayed using either the color or the area of the hexagons.

```python
# FIRST CREATE A DATAFRAME OF RANDOM VALUES
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])

# MAKE A HEXAGONAL BIN PLOT
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges');
```
